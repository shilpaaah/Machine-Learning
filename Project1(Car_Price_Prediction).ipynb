```python
# Encoding: Get_dummies
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1
```

Now we have to join df1 and df using concatenate function.


```python
df2=pd.concat([df,df1],axis=1)
df2
```

Now we drop unwanted columns. ==> name,unnamed:0,new_price, and also the object columns=> bcz we have their encoded version.

```python
df2.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Fuel_Type_Electric'],axis=1,inplace=True)
df2
```

Now we have to replce the string value from the columns 'Mileage','engine','Power'  for that==>

```python
df2['Mileage']=df2['Mileage'].str.replace('km/kg','')
df2['Mileage']=df2['Mileage'].str.replace('kmpl','')
df2['Engine']=df2['Engine'].str.replace('CC','')
df2['Power']=df2['Power'].str.replace('bhp','')
df2
```

null should be also replaced as 0 in the given data : in Mileage and power null value is typed it should be replaced with 0==>'0'

```python
#df2['Engine']=df2['Engine'].str.replace('null','0')
df2['Power']=df2['Power'].str.replace('null','0')
df2['Mileage']=df2['Mileage'].str.replace('null','0')
df2
```

```python
df2.dtypes
```

Still their dtypes haven't changed so we have to change the datatypes also.

```python
df2['Engine']=df2['Engine'].astype(float)
df2['Mileage']=df2['Mileage'].astype(float)
df2['Power']=df2['Power'].astype(float)
df2.dtypes
```

```python
df2.isna().sum()
```

Null value was converted to '0' and then to integer dtype.==> There will be not an engine or power with value 0. So we have to treat that '0' as missing value===NaN.  and  This is handled using Num Py.

```python
df2.loc[df2.Mileage==0,'Mileage']=np.NaN
df2.loc[df2.Power==0,'Power']=np.NaN
df2.isna().sum()
```

Now we have to fill the missing values using mean . and seat capacity using mode as it cant be float.

```python
df2['Mileage']=df2['Mileage'].fillna(df2['Mileage'].mean())
df2['Engine']=df2['Engine'].fillna(df2['Engine'].mean())
df2['Power']=df2['Power'].fillna(df2['Power'].mean())
df2['Seats']=df2['Seats'].fillna(df2['Seats'].mode()[0])
df2
```

```python
df2.isna().sum()
```
