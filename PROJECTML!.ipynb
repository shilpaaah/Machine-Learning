### **Data Cleaning and Preprocessing**
#### Handling missing values

```python
#Replace missing values in 'type', 'body', 'acidity', and 'year' columns with appropriate strategies
# Data Cleaning and Preprocessing
# Handling Missing Values
df['type'].fillna(df['type'].mode()[0], inplace=True)
df['body'].fillna(df['body'].median(), inplace=True)
df['acidity'].fillna(df['acidity'].median(), inplace=True)
df['year'] = pd.to_numeric(df['year'], errors='coerce')  # Convert 'year' to numeric, handle errors by setting them to NaN
df['year'].fillna(df['year'].median(), inplace=True)
```

### **Encoding**

```python
# Encoding object features
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['type'] = label_encoder.fit_transform(df['type'])
df['country'] = label_encoder.fit_transform(df['country'])
df['region'] = label_encoder.fit_transform(df['region'])
df['winery'] = label_encoder.fit_transform(df['winery'])
df['wine'] = label_encoder.fit_transform(df['wine'])
df['year'] = label_encoder.fit_transform(df['year'])
```

```python
# Grid search cv
from sklearn.model_selection import GridSearchCV
```

#### **features and label**

```python
# Feature selection
features = ['wine','year','region','price','type','body','acidity','num_reviews','winery']
x=df[features]
x.ndim
```

```python
y=df['rating']
y.ndim
```

### **Splitting the data into training and testing sets**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

### **Model Creation and Hyperparameter tunning**

```python
# Define models and their respective parameter grids
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
models = {'LinearRegression': (LinearRegression(), {}),
    'RandomForestRegressor': (RandomForestRegressor(random_state=42), {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]}),
    'DecisionTreeRegressor': (DecisionTreeRegressor(random_state=42), {
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]})}

# Perform hyperparameter tuning for each model
for name, (model, param_grid) in models.items():
    print(f"Hyperparameter tuning for {name}:")
    grid_search = GridSearchCV(model, param_grid, cv=10, scoring='neg_mean_squared_error')
    grid_search.fit(x_train, y_train)

    # Display the best parameters and their corresponding score
    print("Best Parameters:", grid_search.best_params_)
```

```python
# Training the model with the best parameters
l_model=LinearRegression()
r_model=RandomForestRegressor(max_depth=10,min_samples_leaf=1,min_samples_split=2,n_estimators=200)
d_model=DecisionTreeRegressor(max_depth=10,min_samples_leaf=4, min_samples_split=10)
lst_model=[l_model,r_model,d_model]
```

```python
for i in lst_model:
  print("Model name is",i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('Squared error is',mean_squared_error(y_test,y_pred))
  print('r2_score is',r2_score(y_test,y_pred))
  print('********************************************************************')
```

### **Evaluating the model**

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')  # Plotting the diagonal line
plt.title('Linear Regression: y_test vs y_pred')
plt.xlabel('Actual Rating (y_test)')
plt.ylabel('Predicted Rating (y_pred)')
plt.grid(True)
plt.show()
```
