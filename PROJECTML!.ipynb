```python
#Replace missing values in 'type', 'body', 'acidity', and 'year' columns with appropriate strategies
# Data Cleaning and Preprocessing
# Handling Missing Values
df['type'].fillna(df['type'].mode()[0], inplace=True)
df['body'].fillna(df['body'].median(), inplace=True)
df['acidity'].fillna(df['acidity'].median(), inplace=True)
df['year'] = pd.to_numeric(df['year'], errors='coerce')  # Convert 'year' to numeric, handle errors by setting them to NaN
df['year'].fillna(df['year'].median(), inplace=True)
```

### **Removing Outliers Using Isolation tree**

```python
# Removing outliers using Isolation Forest
from sklearn.ensemble import IsolationForest
outlier_detector=IsolationForest(contamination=0.05, random_state=42)
df['outlier'] = outlier_detector.fit_predict(df[['price']])
data = df[df['outlier']==1]
```

outlier.jpg  

### Using Isolation Forest for Outlier Detection:
* Isolation Forest is employed to identify and remove outliers in the 'price' feature, ensuring a more robust and accurate analysis.
* Outliers can significantly impact statistical measures, and their removal improves the reliability of subsequent visualizations and modeling.
* The contamination parameter is set to 5%, allowing the Isolation Forest to isolate and eliminate the top 5% of extreme values in the 'price' distribution.
* This step enhances the overall quality of the dataset for downstream analysis and modeling.






### **Encoding**

```python
# Encoding object features
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['type'] = label_encoder.fit_transform(df['type'])
df['country'] = label_encoder.fit_transform(df['country'])
df['region'] = label_encoder.fit_transform(df['region'])
df['winery'] = label_encoder.fit_transform(df['winery'])
df['wine'] = label_encoder.fit_transform(df['wine'])
df['year'] = label_encoder.fit_transform(df['year'])
```

#### **features and label**

```python
# Feature selection
features = ['winery','wine','year','num_reviews','country','region','price','type','body','acidity']
x=df[features]
x.ndim
```

```python
y=df['acidity']
y.ndim
```

### **Splitting the data into training and testing sets**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

### **Standardizing the features.**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

### **Model Creation and Hyperparameter tunning**

```python
# Model creation and hyperparameter tuning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn=KNeighborsClassifier()
param_grid={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
grid_search= GridSearchCV(knn,param_grid,cv=10, scoring='accuracy')
grid_search.fit(x_train,y_train)
```

```python
# Display the best parameters
print("Best Parameters:",grid_search.best_params_)
```

```python
# Training the model with the best parameters
model=KNeighborsClassifier(n_neighbors=3,weights='uniform')
model.fit(x_train,y_train)
# Prediction
y_pred=model.predict(x_test)
y_pred
```

### **Evaluating the model**

```python
# Evaluating the model accuracy score and classification _report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)

```

```python
# Classification report
print("Classification Report:\n", classification_report(y_test,y_pred))
```

#### **Visualization of Accuracy**

```python
plt.figure(figsize=(8, 6))
sns.heatmap([[accuracy]], annot=True, cmap='Blues', fmt='.2%')
plt.title('Accuracy Visualization')
plt.show()
```
