```python
# Temperature Pressure
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('/content/poly.xlsx')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
x=df['Temperature'].tolist()
y=df['Pressure'].tolist()
plt.scatter(x,y)
plt.xlabel('Temperture')
plt.ylabel('Pressure')
plt.title('Temp Vs Pressure')
plt.xticks(x)
plt.yticks(y)
```

```python
x=df.iloc[:,1:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#model Creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
y
```

```python
# Regression line plotting:
plt.scatter(x,y,color='r')
plt.plot(x,y_pred,color='g')
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('MAPE=',mean_absolute_percentage_error(y,y_pred))
```

```python
print('r2 score=',r2_score(y,y_pred))
```

```python
# Implementing Polynomial Regression in this :
from sklearn.preprocessing import PolynomialFeatures
x_poly=PolynomialFeatures(degree=3)
x_feature=x_poly.fit_transform(x)
x_feature
```

```python
# Model Creation:
model1=LinearRegression()
model1.fit(x_feature,y)
y_poly=model1.predict(x_feature)
y_poly
```

```python
# Graph plotting:
plt.scatter(x,y,color='r')
plt.plot(x,y_poly,color='g')
plt.title('Temp Vs Pressure',color='b',fontsize=20)
plt.xlabel('Temperature',fontsize=15)
plt.ylabel('Pressure',fontsize=15)

```

```python
# Evaluation:
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('MAPE=',mean_absolute_percentage_error(y,y_poly))
print('r2 score=',r2_score(y,y_poly))
```

```python
# DF : Actual value  Simple_Regression value    Poly_value
df=pd.DataFrame({'Actual_value':y,'Simple':y_pred,'Poly':y_poly})
df
```
